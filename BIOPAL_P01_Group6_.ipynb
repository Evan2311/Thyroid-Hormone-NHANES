## ❓ Research Question and Objectives

### 5.1 Original Research Question

> Among adults aged 18 and above from 2007–2012, does demographic and clinical factors—such as **BMI**, **age**, **gender**, **alcohol use**, **oral health**, and **physical activity**—affect thyroid function as indicated by **TSH**, **T3**, and **Free T4** levels?

---

### 5.2 Justification and Literature Review

Thyroid dysfunction is influenced by numerous variables, but most studies examine them **in isolation**. This study is **novel** in integrating multiple predictors and assessing their **combined impact** on three core thyroid biomarkers.

#### 📚 Supporting Literature:

- **BMI and Age**: High BMI and older age are recognized predictors of hypothyroidism (Hollowell et al., 2002). However, **few studies explore their interaction with other lifestyle factors**.
  
- **Physical Activity**: While exercise may support thyroid regulation, evidence is **mixed** and often **fails to control** for confounders like BMI or psychological stress (Brent, 2012).

- **Alcohol Use**: Chronic alcohol consumption is linked to **reduced T3 and T4 levels**, due to both thyroid gland suppression and altered **peripheral metabolism** (Völzke et al., 2006).

- **Sex Differences**: Women are **more prone** to thyroid disorders. Yet, **few studies stratify** analyses by sex or test for **interaction effects** (Taylor et al., 2018).

- **Oral Health**: Poor oral hygiene can lead to **inflammation and autoimmune activity**, possibly affecting the thyroid. However, **large-scale data** on this is limited (Nazik et al., 2017).

- **Integrated Models Are Rare**: Most studies look at **single-factor relationships**. This research builds a **multi-factorial model** using a large, representative dataset to study **TSH, T3, and T4** collectively.

```python
import pandas as pd
import numpy as np

# -----------------------------
# Step 0: Define years and NHANES cycles
# -----------------------------
years = ["2007","2009","2011"]
cycles = ["E","F","G"]

# -----------------------------
# Step 1: Load All Raw Datasets
# -----------------------------
print("Loading all raw datasets...")

def load_data(file_prefix, columns):
    urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/{file_prefix}_{c}.xpt"
            for y, c in zip(years, cycles)]
    df_list = []
    for url in urls:
        try:
            df = pd.read_sas(url)
            existing_cols = [col for col in columns if col in df.columns]
            if existing_cols:
                df_list.append(df[existing_cols])
                print(f"Loaded {url} with columns: {existing_cols}")
            else:
                print(f"No matching columns found in {url}")
        except Exception as e:
            print(f"Could not load or process {url}. Error: {e}")
    if df_list:
        return pd.concat(df_list, ignore_index=True)
    else:
        print(f"No data loaded for {file_prefix}. Returning empty DataFrame.")
        return pd.DataFrame()

df_demo = load_data("DEMO", ["SEQN","SDDSRVYR", "RIAGENDR", "RIDAGEYR"])
df_bmi = load_data("BMX", ["SEQN", "SDDSRVYR","BMXBMI"])
df_physical = load_data("PAQ", ["SEQN", "SDDSRVYR","PAD680"])
df_drinking = load_data("ALQ", ["SEQN", "SDDSRVYR", "ALQ120Q","ALQ151"])
df_thyroid = load_data("THYROD",["SEQN","SDDSRVYR","LBXTSH1","LBXT4F","LBXT3F"])

print("Loading OHAPOS data from OHX and OHXREF...")

# Each tuple: (year, cycle, file_prefix, SDDSRVYR value)
oral_files = [
    ("2007", "E", "OHX", 5),
    ("2009", "F", "OHXREF", 6),
    ("2011", "G", "OHXREF", 7)
]

oral_dfs = []
for year, cycle, prefix, survey_cycle in oral_files:
    url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{year}/DataFiles/{prefix}_{cycle}.xpt"
    try:
        df = pd.read_sas(url)
        ohapos_col = next((col for col in df.columns if "OHAPOS" in col.upper()), None)
        if ohapos_col:
            df = df[["SEQN", ohapos_col]].copy()
            df["SDDSRVYR"] = survey_cycle  # Manually add SDDSRVYR
            df.rename(columns={ohapos_col: "OHAPOS"}, inplace=True)
            oral_dfs.append(df)
            print(f"Loaded {url} with OHAPOS column: {ohapos_col}")
        else:
            print(f"OHAPOS column not found in {url}")
    except Exception as e:
        print(f"Could not load {url}. Error: {e}")

# Combine oral health data
if oral_dfs:
    df_oral = pd.concat(oral_dfs, ignore_index=True)
else:
    df_oral = pd.DataFrame()
    print("No OHAPOS data loaded.")
```

```python
# ---------------------------------------------------------------------------------
# Step 3: Load and DEFINITIVELY Standardize Alcohol History (FINAL CORRECTED LOGIC)
# ---------------------------------------------------------------------------------
print("Loading and standardizing alcohol data...")
#ALQ120Q - How often drink alcohol over past 12 mos

def drinking_category(row):
    freq = row["ALQ120Q"]
    binge = row["ALQ151"]

    # Handle invalid/missing data
    if pd.isna(freq) or freq in [77, 99] or pd.isna(binge) or binge in [7, 9]:
        return np.nan

    # Never drank in the past year
    if freq == 7:
        return "Non-drinker"

    # Binge drinker: drank + had 4/5+ drinks/day at least once
    if binge == 1:
        return "Binge drinker"

    # Frequency-based categories
    if freq in [1, 2]:
        return "Heavy drinker"
    elif freq in [3, 4]:
        return "Moderate drinker"
    elif freq in [5, 6]:
        return "Light drinker"

    return np.nan


df_drinking["Drinking habits"] = df_drinking.apply(drinking_category, axis=1)

#df_drinking.drop(columns=["ALQ120Q", "ALQ151"], inplace=True)


#Step 4: Load in sedentary activity

# -----------------------------
# Define years and NHANES cycles

# Load PAD680 data (2007–2012)
years = ["2007", "2009", "2011"]
cycles = ["E", "F", "G"]
df_physical = load_data("PAQ", ["SEQN", "SDDSRVYR", "PAD680"])

# Apply classification

def sedentary_behavior_pad680(row):
    val = row["PAD680"]
    if pd.isna(val) or val in [7777, 9999]:
        return np.nan
    elif val >= 420:
        return "Sedentary"
    elif 180 <= val < 420:
        return "Active"
    elif val < 180:
        return "Very active"
    return np.nan

# Assign category
df_physical["Activity_Level"] = df_physical.apply(sedentary_behavior_pad680, axis=1)

#df_physical.drop(columns=["PAD680"], inplace=True)

print (df_physical["Activity_Level"].unique())


#Sedentary behavior

#PAD680 - Minutes of sedentary activity in a week. 2007 - 2018
#Minutes/week Hours/day Category
#≥ 420    ≥ 6 hrs Sedentary
#180–419  3–6 hrs Active
#< 180    < 3 hrs Very Active

# -----------------------------
# Step 5: Classify Oral Health Status
# -----------------------------

def classify_oral_health(val):
    if pd.isna(val) or val in [7, 9]:
        return np.nan
    elif val == 1:
        return "Good oral health"
    elif val == 2:
        return "Poor oral health"
    else:
        return "Unknown"

df_oral["Oral_Health_Status"] = df_oral["OHAPOS"].apply(classify_oral_health)


# -----------------------------
# Step 6a: Classify TSH levels from LBXTSH1
# -----------------------------

def classify_tsh(val):
    if pd.isna(val):
        return np.nan
    elif val < 0.4:
        return "Low TSH (Hyperthyroid)"
    elif val <= 4.0:
        return "Normal TSH"
    else:
        return "High TSH (Hypothyroid)"

# Apply classification
df_thyroid["TSH_Status"] = df_thyroid["LBXTSH1"].apply(classify_tsh)

# -----------------------------
# Step 6b: Classify Free T3 (LBXT3F)
# -----------------------------

def classify_t3f(val):
    if pd.isna(val):
        return np.nan
    elif val < 2.0:
        return "Low Free T3"
    elif val <= 4.4:
        return "Normal Free T3"
    else:
        return "High Free T3"

df_thyroid["Free_T3_Status"] = df_thyroid["LBXT3F"].apply(classify_t3f)

# -----------------------------
# Step 6c: Classify Free T4 (LBXT4F)
# -----------------------------

def classify_t4f(val):
    if pd.isna(val):
        return np.nan
    elif val < 0.8:
        return "Low Free T4"
    elif val <= 1.8:
        return "Normal Free T4"
    else:
        return "High Free T4"

df_thyroid["Free_T4_Status"] = df_thyroid["LBXT4F"].apply(classify_t4f)


# -----------------------------
# Step 8: Merge all datasets
# -----------------------------

print("Merging all datasets...")
df = df_demo.merge(df_bmi, on="SEQN", how="left") \
            .merge(df_demo, on="SEQN", how="left") \
            .merge(df_oral, on="SEQN", how="left") \
            .merge(df_physical, on="SEQN", how="left") \
            .merge(df_drinking, on="SEQN", how="left") \
            .merge(df_thyroid, on="SEQN",how="left")
print(df.isnull().sum()) #Check for missing data

# -----------------------------
# Step 9: Final Cleaning and Filtering
# -----------------------------
print("Performing final cleaning and filtering...")
#df.drop(columns=[
#    "SMQ020", "SMQ040", 'EverDrinkerLife', 'DrankPastYear', 'DrinksPerDay',
#    "PAD680", "HIQ210", "INDFMPIR", "MCQ220", "SEQN", "RIDRETH1", "HID010" #Removed RIDRETH1 and HID010 after mapping
#], inplace=True, errors='ignore')
df.replace("Unknown", np.nan, inplace=True)

#Map years
cycle_map = {
    5: '2007–2008', 6: '2009–2010', 7: '2011–2012'
}
df['CycleYears'] = df['SDDSRVYR'].map(cycle_map)
display(df["CycleYears"])


# -----------------------------
# Step 10: Final Output
# -----------------------------
print("\n--- FINAL OUTPUT AFTER ALL CORRECTIONS AND FILTERING ---")
print("\nOral Health Status counts:")
print(df["Oral_Health_Status"].value_counts(dropna=False))
print("\nAlcohol Use counts:")
print(df["Drinking habits"].value_counts(dropna=False))
print("\nThyroid Stim. Hormone counts:")
print(df["TSH_Status"].value_counts(dropna=False))
print("\nFree T4 counts:")
print(df["Free_T4_Status"].value_counts(dropna=False))
print("\nFree T3 counts:")
print(df["Free_T3_Status"].value_counts(dropna=False))
print("\nActivity Level counts")
print(df["Activity_Level"].value_counts(dropna=False))
```

```python
import pandas as pd
import numpy as np

# -----------------------------
# Step 0: Define years and NHANES cycles
# -----------------------------
years = ["2007","2009","2011"]
cycles = ["E","F","G"]

# -----------------------------
# Step 1: Load All Raw Datasets
# -----------------------------
print("Loading all raw datasets...")

def load_data(file_prefix, columns):
    urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/{file_prefix}_{c}.xpt"
            for y, c in zip(years, cycles)]
    df_list = []
    for url in urls:
        try:
            df = pd.read_sas(url)
            existing_cols = [col for col in columns if col in df.columns]
            if existing_cols:
                df_list.append(df[existing_cols])
                print(f"Loaded {url} with columns: {existing_cols}")
            else:
                print(f"No matching columns found in {url}")
        except Exception as e:
            print(f"Could not load or process {url}. Error: {e}")
    if df_list:
        return pd.concat(df_list, ignore_index=True)
    else:
        print(f"No data loaded for {file_prefix}. Returning empty DataFrame.")
        return pd.DataFrame()

df_demo = load_data("DEMO", ["SEQN","SDDSRVYR", "RIAGENDR", "RIDAGEYR"])
df_bmi = load_data("BMX", ["SEQN", "SDDSRVYR","BMXBMI"])
df_physical = load_data("PAQ", ["SEQN", "SDDSRVYR","PAD680"])
df_drinking = load_data("ALQ", ["SEQN", "SDDSRVYR", "ALQ120Q","ALQ151"])
df_thyroid = load_data("THYROD",["SEQN","SDDSRVYR","LBXTSH1","LBXT4F","LBXT3F"])

print("Loading OHAPOS data from OHX and OHXREF...")

# Each tuple: (year, cycle, file_prefix, SDDSRVYR value)
oral_files = [
    ("2007", "E", "OHX", 5),
    ("2009", "F", "OHXREF", 6),
    ("2011", "G", "OHXREF", 7)
]

oral_dfs = []
for year, cycle, prefix, survey_cycle in oral_files:
    url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{year}/DataFiles/{prefix}_{cycle}.xpt"
    try:
        df = pd.read_sas(url)
        ohapos_col = next((col for col in df.columns if "OHAPOS" in col.upper()), None)
        if ohapos_col:
            df = df[["SEQN", ohapos_col]].copy()
            df["SDDSRVYR"] = survey_cycle  # Manually add SDDSRVYR
            df.rename(columns={ohapos_col: "OHAPOS"}, inplace=True)
            oral_dfs.append(df)
            print(f"Loaded {url} with OHAPOS column: {ohapos_col}")
        else:
            print(f"OHAPOS column not found in {url}")
    except Exception as e:
        print(f"Could not load {url}. Error: {e}")

# Combine oral health data
if oral_dfs:
    df_oral = pd.concat(oral_dfs, ignore_index=True)
else:
    df_oral = pd.DataFrame()
    print("No OHAPOS data loaded.")

# -----------------------------
# Step 8: Merge all datasets
# -----------------------------
print("Merging all datasets...")
# Use SEQN for merging. Handle potential duplicate column names from merging (e.g., SDDSRVYR)
df_merged = df_demo.merge(df_bmi.drop(columns=['SDDSRVYR'], errors='ignore'), on="SEQN", how="left") \
                   .merge(df_oral.drop(columns=['SDDSRVYR'], errors='ignore'), on="SEQN", how="left") \
                   .merge(df_physical.drop(columns=['SDDSRVYR'], errors='ignore'), on="SEQN", how="left") \
                   .merge(df_drinking.drop(columns=['SDDSRVYR'], errors='ignore'), on="SEQN", how="left") \
                   .merge(df_thyroid.drop(columns=['SDDSRVYR'], errors='ignore'), on="SEQN", how="left")

# Keep only one SDDSRVYR column (from df_demo)
df_subset = df_merged.copy()

print(df_subset.isnull().sum()) #Check for missing data in the raw merged df

# -----------------------------
# Step 6: Classify Variables
# -----------------------------
print("Classifying variables...")

# 1. Classify Free T4 levels (LBXT4F)
def classify_t4f(val):
    if pd.isna(val):
        return np.nan
    elif val < 0.4:
        return "Low T4F (Hypothyroid)"
    elif val <= 4.0:
        return "Normal T4F"
    else: # val > 4.0
        return "High T4F (Hyperthyroid)"

df_subset['T4F_Status'] = df_subset['LBXT4F'].apply(classify_t4f)

# 2. Classify Sedentary Activity (PAD680)
def sedentary_behavior_pad680(val):
    if pd.isna(val) or val in [7777, 9999]:
        return np.nan
    elif val < 180:
        return "Very active"
    elif 180 <= val < 420:
        return "Active"
    elif val >= 420:
        return "Sedentary"
    return np.nan # Should not be reached with valid input

df_subset['Activity_Level'] = df_subset['PAD680'].apply(sedentary_behavior_pad680)

# 3. Classify BMI (BMXBMI)
def classify_bmi(bmi):
    if pd.isna(bmi):
        return np.nan
    elif bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal weight'
    elif 25 <= bmi < 30:
        return 'Overweight'
    elif 30 <= bmi < 35:
        return 'Obese I'
    elif 35 <= bmi < 40:
        return 'Obese II'
    else:
        return 'Obese III'

df_subset['BMI_Category'] = df_subset['BMXBMI'].apply(classify_bmi)

# 4. Classify Age (RIDAGEYR)
def classify_age(age):
    if pd.isna(age):
        return np.nan
    elif age < 13:
        return 'Child'
    elif 13 <= age < 20:
        return 'Teen'
    elif 20 <= age < 65:
        return 'Adult'
    else:
        return 'Senior'

df_subset['AgeGroup'] = df_subset['RIDAGEYR'].apply(classify_age)

# 5. Classify Drinking Habits (ALQ120Q, ALQ151)
def drinking_category(row):
    freq = row["ALQ120Q"]
    binge = row["ALQ151"]

    # Handle invalid/missing data
    if pd.isna(freq) or freq in [77, 99] or pd.isna(binge) or binge in [7, 9]:
        return np.nan

    # Never drank in the past year
    if freq == 7:
        return "Non-drinker"

    # Binge drinker: drank + had 4/5+ drinks/day at least once
    if binge == 1:
        return "Binge drinker"

    # Frequency-based categories
    if freq in [1, 2]:
        return "Heavy drinker"
    elif freq in [3, 4]:
        return "Moderate drinker"
    elif freq in [5, 6]:
        return "Light drinker"

    return np.nan

df_subset["Drinking habits"] = df_subset.apply(drinking_category, axis=1)

# 6. Map Gender (RIAGENDR)
gender_map = {1.0: 'Male', 2.0: 'Female'}
df_subset['RIAGENDR'] = df_subset['RIAGENDR'].map(gender_map)
# Missing values will become NaN by default using map

# 7. Classify Oral Health Status (OHAPOS)
def classify_oral_health(value):
    if pd.isna(value) or value in [7, 9]: # Assuming 7, 9 are also invalid/unknown codes
        return np.nan # Returning NaN for unknown/missing based on previous steps' pattern
    elif value == 1: # Example: Assuming 1 is 'Good' based on typical coding
        return 'Good oral health'
    elif value == 2: # Example: Assuming 2 is 'Poor'
         return 'Poor oral health'
    else:
        # Handle any other values if necessary based on data dictionary
        return 'Unknown' # Treat other codes as unknown/missing

df_subset['Oral_Health_Status'] = df_subset['OHAPOS'].apply(classify_oral_health)
df_subset.replace("Unknown", np.nan, inplace=True) # Replace 'Unknown' with NaN after classification


# 8. Map Survey Years (SDDSRVYR)
cycle_map = {5.0: '2007–2008', 6.0: '2009–2010', 7.0: '2011–2012'}
df_subset['CycleYears'] = df_subset['SDDSRVYR'].map(cycle_map)
# Missing values will become NaN by default using map

# 9. Classify TSH and Free T3 status
def classify_tsh(val):
    if pd.isna(val):
        return np.nan
    elif val < 0.4:
        return "Low TSH (Hypothyroid)"
    elif val <= 4.0:
        return "Normal TSH"
    else: # val > 4.0
        return "High TSH (Hyperthyroid)"

def classify_t3(val):
    if pd.isna(val):
        return np.nan
    elif val < 2.0:
        return "Low Free T3"
    elif val <= 4.4:
        return "Normal Free T3"
    else: # val > 4.4
        return "High Free T3"

df_subset['TSH_Status'] = df_subset['LBXTSH1'].apply(classify_tsh)
df_subset['Free_T3_Status'] = df_subset['LBXT3F'].apply(classify_t3)


# Drop raw data columns after classification
df_subset.drop(columns=[
    'OHAPOS', 'PAD680', 'ALQ120Q', 'ALQ151', 'LBXTSH1', 'LBXT4F', 'LBXT3F' # Added LBXT3F and LBXTSH1
], errors='ignore', inplace=True)


# -----------------------------
# Step 9: Final Output (Using df_subset consistently)
# -----------------------------
print("\n--- FINAL OUTPUT AFTER ALL CORRECTIONS AND FILTERING ---")
print("\nOral Health Status counts:")
print(df_subset["Oral_Health_Status"].value_counts(dropna=False))
print("\nAlcohol Use counts:")
print(df_subset["Drinking habits"].value_counts(dropna=False))
print("\nFree Thyroxine counts:")
print(df_subset["T4F_Status"].value_counts(dropna=False)) # Using T4F_Status as created
print("\nActivity Level counts")
print(df_subset["Activity_Level"].value_counts(dropna=False))
print("\nCycle Years counts:")
print(df_subset["CycleYears"].value_counts(dropna=False)) # Verify CycleYears
print("\nTSH Status counts:")
print(df_subset["TSH_Status"].value_counts(dropna=False)) # Verify TSH_Status
print("\nFree T3 Status counts:")
print(df_subset["Free_T3_Status"].value_counts(dropna=False)) # Verify Free_T3_Status
```

### 1. Descriptive Statistics

For all **continuous variables** (e.g. age, BMI, hormone levels):
- **Mean**
- **Median**
- **Standard Deviation (SD)**
- **Quartiles (Q1, Q3)**
- **Skewness**

For all **categorical variables** (e.g. gender, oral health status, thyroid classification):
- Display frequency counts and proportions.
- Identify dominant and underrepresented categories.

### 3. Normality Checks

To assess whether continuous variables follow a normal distribution:
- **Shapiro-Wilk Test**
- **Histograms**
- **Q-Q Plots**

Variables failing normality tests may require transformation or non-parametric tests.

---



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels.api as sm
import warnings
warnings.filterwarnings('ignore')

# Load XPT file
file_path = "/content/THYROD_E.xpt"
df = pd.read_sas(file_path)

# Preview data
print("Data shape:", df.shape)
print("Columns:", df.columns.tolist())

# -------------------------------
# Descriptive statistics (continuous variables)
# -------------------------------
cont_vars = df.select_dtypes(include=[np.number]).columns

desc_stats = df[cont_vars].describe().T
desc_stats["skewness"] = df[cont_vars].skew()

print("\nDescriptive Statistics:\n", desc_stats)

# -------------------------------
# Frequency tables (categorical variables)
# -------------------------------
cat_vars = df.select_dtypes(include=['category', 'object']).columns

for col in cat_vars:
    print(f"\nFrequency Table for {col}:\n")
    print(df[col].value_counts(dropna=False))

# -------------------------------
# Normality checks (Shapiro-Wilk, Histograms, Q-Q plots)
# -------------------------------
from scipy.stats import shapiro

for col in cont_vars:
    col_data = df[col].dropna()
    if len(col_data) >= 3 and len(col_data) <= 5000:  # Shapiro limit
        stat, p = shapiro(col_data)
        print(f"\nShapiro-Wilk Test for {col} — W={stat:.4f}, p={p:.4f}")

        # Histogram
        sns.histplot(col_data, kde=True)
        plt.title(f"Histogram of {col}")
        plt.xlabel(col)
        plt.ylabel("Frequency")
        plt.show()

        # Q-Q plot
        sm.qqplot(col_data, line='s')
        plt.title(f"Q-Q Plot for {col}")
        plt.show()

# -------------------------------
# Boxplots and Visuals for Key Variables
# -------------------------------
key_vars = ['LBXT4FS']  # You can add more if needed

for var in key_vars:
    if var in df.columns:
        # Boxplot
        sns.boxplot(x=df[var])
        plt.title(f"Boxplot of {var}")
        plt.show()

        # Histogram
        sns.histplot(df[var], kde=True)
        plt.title(f"Histogram of {var}")
        plt.xlabel(var)
        plt.show()

        # If it's categorical, show bar chart
        if df[var].dtype == 'object' or df[var].dtype.name == 'category':
            df[var].value_counts().plot(kind='bar')
            plt.title(f"Bar Chart of {var}")
            plt.ylabel("Frequency")
            plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# File paths
files = {
    "2007-2008": "/content/THYROD_E.xpt",
    "2009-2010": "/content/THYROD_F.xpt",
    "2011-2012": "/content/THYROD_G.xpt"
}

# Load and merge datasets
dfs = []
for year, path in files.items():
    df = pd.read_sas(path)
    df['Survey_Year'] = year
    dfs.append(df)

df_all = pd.concat(dfs, ignore_index=True)

# Define marker classification functions
def classify_t4(val):
    if val < 0.8:
        return 'Low'
    elif val > 1.8:
        return 'High'
    else:
        return 'Normal'

def classify_t3(val):
    if val < 2.0:
        return 'Low'
    elif val > 4.4:
        return 'High'
    else:
        return 'Normal'

def classify_tsh(val):
    if val < 0.4:
        return 'Low'
    elif val > 4.0:
        return 'High'
    else:
        return 'Normal'

# Marker configuration
markers = {
    "LBXT4F": {
        "label": "Free T4 (ng/dL)",
        "classifier": classify_t4
    },
    "LBXT3F": {
        "label": "Free T3 (pg/mL)",
        "classifier": classify_t3
    },
    "LBXTSH1": {
        "label": "TSH (mIU/L)",
        "classifier": classify_tsh
    }
}

# Process each marker
for marker, config in markers.items():
    if marker in df_all.columns:
        df_marker = df_all[[marker, "Survey_Year"]].dropna()
        df_marker["Category"] = df_marker[marker].apply(config["classifier"])

        # Median
        median = df_marker.groupby("Survey_Year")[marker].median().reset_index(name="Median")

        # % Abnormal (Low + High)
        abnormal = df_marker.groupby("Survey_Year")["Category"].apply(
            lambda x: (x.isin(["Low", "High"])).sum() / len(x) * 100
        ).reset_index(name="Abnormal (%)")

        # Merge results
        trend_df = pd.merge(median, abnormal, on="Survey_Year")
        print(f"\n{config['label']} Trends:\n", trend_df)

        # Plot
        plt.figure(figsize=(10, 6))
        sns.lineplot(data=trend_df, x="Survey_Year", y="Median", marker="o", label=f"Median {config['label']}")
        sns.lineplot(data=trend_df, x="Survey_Year", y="Abnormal (%)", marker="o", label=f"% Abnormal {config['label']}")

        plt.title(f"NHANES Trends: {config['label']} (2007–2012)")
        plt.ylabel("Value / Percent")
        plt.xticks(rotation=45)
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()
    else:
        print(f"⚠️ Marker {marker} not found in dataset.")
```

```python
import zipfile
import os
import pandas as pd
import numpy as np

# -----------------------------
# Step 0: Define zip file path and extraction directory
# -----------------------------
zip_file_path = "/content/New Compressed (zipped) Folder.zip"
extract_dir = "/content/extracted_nhanes_data"

# Create the extraction directory if it doesn't exist
os.makedirs(extract_dir, exist_ok=True)

# Unzip the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

# -----------------------------
# Step 1: Define file prefixes and columns
# -----------------------------
print("Defining file prefixes and columns...")

# Dictionary mapping file prefixes to columns to load
file_column_map = {
    "DEMO": ["SEQN", "SDDSRVYR", "RIAGENDR", "RIDAGEYR"],
    "BMX": ["SEQN", "SDDSRVYR", "BMXBMI"],
    "PAQ": ["SEQN", "SDDSRVYR", "PAD680"],
    "ALQ": ["SEQN", "SDDSRVYR", "ALQ120Q", "ALQ151"],
    "THYROD": ["SEQN", "SDDSRVYR", "LBXTSH1", "LBXT4F", "LBXT3F"],
}

# -----------------------------
# Step 6: Create load_data function
# -----------------------------
print("Creating load_data function...")

def load_data(file_prefix, columns):
    df_list = []
    # List all files in the extracted directory that match the file_prefix
    # Handle potential duplicated files with numbers in parentheses
    extracted_files = [os.path.join(extract_dir, f) for f in os.listdir(extract_dir)
                       if f.startswith(file_prefix) and f.endswith('.xpt')]

    if not extracted_files:
        print(f"No files found for prefix {file_prefix} in {extract_dir}")
        return pd.DataFrame()

    for file_path in extracted_files:
        try:
            df = pd.read_sas(file_path)
            existing_cols = [col for col in columns if col in df.columns]
            if existing_cols:
                df_list.append(df[existing_cols])
                print(f"Loaded {file_path} with columns: {existing_cols}")
            else:
                print(f"No matching columns found in {file_path}")
        except Exception as e:
            print(f"Could not load or process {file_path}. Error: {e}")
    if df_list:
        return pd.concat(df_list, ignore_index=True)
    else:
        print(f"No data loaded for {file_prefix}. Returning empty DataFrame.")
        return pd.DataFrame()

# -----------------------------
# Step 7, 8, 9, 10: Load and process Oral Health (OHX/OHXREF) data
# -----------------------------
print("Loading and processing OHAPOS data from OHX and OHXREF...")

# Each tuple: (year, cycle, file_prefix, SDDSRVYR value)
oral_files = [
    ("2007", "E", "OHX", 5),
    ("2009", "F", "OHXREF", 6),
    ("2011", "G", "OHXREF", 7)
]

oral_dfs = []
for year, cycle, prefix, survey_cycle in oral_files:
    # Construct the expected filename based on prefix and cycle
    # Assuming filenames are like OHX_E.xpt, OHXREF_F.xpt, OHXREF_G.xpt
    file_name = f"{prefix}_{cycle}.xpt"
    file_path = os.path.join(extract_dir, file_name)

    # Also check for files with numbers in parentheses like "OHX_E (1).xpt"
    potential_files = [f for f in os.listdir(extract_dir) if f.startswith(f"{prefix}_{cycle}")]
    found_file = None
    for pf in potential_files:
        if pf.endswith('.xpt'):
            found_file = os.path.join(extract_dir, pf)
            break

    if not found_file:
        print(f"File not found for {prefix}_{cycle}.xpt in {extract_dir}")
        continue

    try:
        df = pd.read_sas(found_file)
        # Find OHAPOS column case-insensitively
        ohapos_col = next((col for col in df.columns if "OHAPOS" in col.upper()), None)
        if ohapos_col:
            # Select SEQN and the found OHAPOS column
            df = df[["SEQN", ohapos_col]].copy()
            df["SDDSRVYR"] = survey_cycle  # Manually add SDDSRVYR
            df.rename(columns={ohapos_col: "OHAPOS"}, inplace=True)
            oral_dfs.append(df)
            print(f"Loaded {found_file} with OHAPOS column: {ohapos_col}")
        else:
            print(f"OHAPOS column not found in {found_file}")
    except Exception as e:
        print(f"Could not load {found_file}. Error: {e}")

# Combine oral health data
if oral_dfs:
    df_oral = pd.concat(oral_dfs, ignore_index=True)
else:
    df_oral = pd.DataFrame()
    print("No OHAPOS data loaded.")


# -----------------------------
# Step 11: Load other datasets
# -----------------------------
print("Loading other datasets...")
df_demo = load_data("DEMO", file_column_map["DEMO"])
df_bmi = load_data("BMX", file_column_map["BMX"])
df_physical = load_data("PAQ", file_column_map["PAQ"])
df_drinking = load_data("ALQ", file_column_map["ALQ"])
df_thyroid = load_data("THYROD", file_column_map["THYROD"])

# -----------------------------
# Step 12: Merge all datasets
# -----------------------------
print("Merging all datasets...")
# Use SEQN for merging. Handle potential duplicate column names from merging (e.g., SDDSRVYR)
df = df_demo.merge(df_bmi, on="SEQN", how="left", suffixes=('_demo', '_bmi')) \
            .merge(df_oral, on="SEQN", how="left") \
            .merge(df_physical, on="SEQN", how="left", suffixes=('_oral', '_physical')) \
            .merge(df_drinking, on="SEQN", how="left", suffixes=('_physical', '_drinking')) \
            .merge(df_thyroid, on="SEQN",how="left", suffixes=('_drinking', '_thyroid'))

# Drop redundant SDDSRVYR columns after merging, keeping one or none if all are NaN
# Find all columns starting with 'SDDSRVYR'
sddsrvyr_cols = [col for col in df.columns if col.startswith('SDDSRVYR')]
if sddsrvyr_cols:
    # Keep the first one and drop the rest
    main_sddsrvyr_col = sddsrvyr_cols[0]
    cols_to_drop = [col for col in sddsrvyr_cols if col != main_sddsrvyr_col]
    df.drop(columns=cols_to_drop, inplace=True, errors='ignore')
    df.rename(columns={main_sddsrvyr_col: 'SDDSRVYR'}, inplace=True) # Rename the kept column

# -----------------------------
# Step 13: Print missing values
# -----------------------------
print("\nMissing values in merged DataFrame:")
print(df.isnull().sum())
```

```python
import pandas as pd
import numpy as np

# 1. Classify Free T4 levels (LBXT4F)
def classify_t4f(val):
    if pd.isna(val):
        return np.nan
    elif val < 0.4:
        return "Low T4F (Hypothyroid)"
    elif val <= 4.0:
        return "Normal T4F"
    else:
        return "High T4F (Hyperthyroid)"

df['T4F_Status'] = df['LBXT4F'].apply(classify_t4f)

# 2. Classify Sedentary Activity (PAD680)
def sedentary_behavior_pad680(val):
    if pd.isna(val) or val in [7777, 9999]:
        return np.nan
    elif val < 180:
        return "Very active"
    elif 180 <= val < 420:
        return "Active"
    elif val >= 420:
        return "Sedentary"
    return np.nan # Should not be reached with valid input

df['Activity_Level'] = df['PAD680'].apply(sedentary_behavior_pad680)

# 3. Classify BMI (BMXBMI)
def classify_bmi(bmi):
    if pd.isna(bmi):
        return np.nan
    elif bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal weight'
    elif 25 <= bmi < 30:
        return 'Overweight'
    elif 30 <= bmi < 35:
        return 'Obese I'
    elif 35 <= bmi < 40:
        return 'Obese II'
    else:
        return 'Obese III'

df['BMI_Category'] = df['BMXBMI'].apply(classify_bmi)

# 4. Classify Age (RIDAGEYR)
def classify_age(age):
    if pd.isna(age):
        return np.nan
    elif age < 13:
        return 'Child'
    elif 13 <= age < 20:
        return 'Teen'
    elif 20 <= age < 65:
        return 'Adult'
    else:
        return 'Senior'

df['AgeGroup'] = df['RIDAGEYR'].apply(classify_age)

# 5. Classify Drinking Habits (ALQ120Q, ALQ151)
def drinking_category(row):
    freq = row["ALQ120Q"]
    binge = row["ALQ151"]

    # Handle invalid/missing data
    if pd.isna(freq) or freq in [77, 99] or pd.isna(binge) or binge in [7, 9]:
        return np.nan

    # Never drank in the past year
    if freq == 7:
        return "Non-drinker"

    # Binge drinker: drank + had 4/5+ drinks/day at least once
    if binge == 1:
        return "Binge drinker"

    # Frequency-based categories
    if freq in [1, 2]:
        return "Heavy drinker"
    elif freq in [3, 4]:
        return "Moderate drinker"
    elif freq in [5, 6]:
        return "Light drinker"

    return np.nan

df["Drinking habits"] = df.apply(drinking_category, axis=1)

# 6. Map Gender (RIAGENDR)
gender_map = {1.0: 'Male', 2.0: 'Female'}
df['RIAGENDR'] = df['RIAGENDR'].map(gender_map)
# Missing values will become NaN by default using map

# 7. Classify Oral Health Status (OHAPOS)
def classify_oral_health(value):
    if pd.isna(value) or value in [7, 9]: # Assuming 7, 9 are also invalid/unknown codes
        return np.nan # Returning NaN for unknown/missing based on previous steps' pattern
    elif value == 1: # Example: Assuming 1 is 'Good' based on typical coding
        return 'Good oral health'
    elif value == 2: # Example: Assuming 2 is 'Poor'
         return 'Poor oral health'
    else:
        # Handle any other values if necessary based on data dictionary
        return np.nan # Treat other codes as unknown/missing

df['Oral_Health_Status'] = df['OHAPOS'].apply(classify_oral_health)

# 8. Map Survey Years (SDDSRVYR)
cycle_map = {5.0: '2007-2008', 6.0: '2009-2010', 7.0: '2011-2012'}
df['SDDSRVYR'] = df['SDDSRVYR'].map(cycle_map)
# Missing values will become NaN by default using map

# 9. Print value counts for verification
print("\nValue Counts for Classified Variables:")
print("\nT4F_Status:")
print(df['T4F_Status'].value_counts(dropna=False))
print("\nActivity_Level:")
print(df['Activity_Level'].value_counts(dropna=False))
print("\nBMI_Category:")
print(df['BMI_Category'].value_counts(dropna=False))
print("\nAgeGroup:")
print(df['AgeGroup'].value_counts(dropna=False))
print("\nDrinking habits:")
print(df['Drinking habits'].value_counts(dropna=False))
print("\nRIAGENDR:")
print(df['RIAGENDR'].value_counts(dropna=False))
print("\nOral_Health_Status:")
print(df['Oral_Health_Status'].value_counts(dropna=False))
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# -----------------------------
# Step 0: Define years and NHANES cycles and file URLs
# -----------------------------
years = ["2007", "2009", "2011"]
cycles = ["E", "F", "G"]

def get_url(year, cycle, file_prefix):
    # Construct the URL for the data file
    return f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{year}/DataFiles/{file_prefix}_{cycle}.xpt"

# -----------------------------
# Step 1: Load All Raw Datasets from URLs
# -----------------------------
print("Loading all raw datasets from URLs...")

def load_data_from_urls(file_prefix, columns):
    df_list = []
    for year, cycle in zip(years, cycles):
        url = get_url(year, cycle, file_prefix)
        try:
            df_temp = pd.read_sas(url)
            existing_cols = [col for col in columns if col in df_temp.columns]
            if existing_cols:
                df_list.append(df_temp[existing_cols])
                print(f"Loaded {url} with columns: {existing_cols}")
            else:
                print(f"No matching columns found in {url}")
        except Exception as e:
            print(f"Could not load or process {url}. Error: {e}")
    if df_list:
        return pd.concat(df_list, ignore_index=True)
    else:
        print(f"No data loaded for {file_prefix}. Returning empty DataFrame.")
        return pd.DataFrame()

df_demo = load_data_from_urls("DEMO", ["SEQN", "SDDSRVYR", "RIAGENDR", "RIDAGEYR"])
df_bmi = load_data_from_urls("BMX", ["SEQN", "SDDSRVYR", "BMXBMI"])
df_physical = load_data_from_urls("PAQ", ["SEQN", "SDDSRVYR", "PAD680"])
df_drinking = load_data_from_urls("ALQ", ["SEQN", "SDDSRVYR", "ALQ120Q", "ALQ151"])
df_thyroid = load_data_from_urls("THYROD", ["SEQN", "SDDSRVYR", "LBXTSH1", "LBXT4F", "LBXT3F"])

print("Loading OHAPOS data from OHX and OHXREF URLs...")

# Each tuple: (year, cycle, file_prefix, SDDSRVYR value)
oral_files_urls = [
    ("2007", "E", "OHX", 5),
    ("2009", "F", "OHXREF", 6),
    ("2011", "G", "OHXREF", 7)
]

oral_dfs = []
for year, cycle, prefix, survey_cycle in oral_files_urls:
    url = get_url(year, cycle, prefix)
    try:
        df_oral_temp = pd.read_sas(url)
        ohapos_col = next((col for col in df_oral_temp.columns if "OHAPOS" in col.upper()), None)
        if ohapos_col:
            df_oral_temp = df_oral_temp[["SEQN", ohapos_col]].copy()
            df_oral_temp["SDDSRVYR"] = survey_cycle  # Manually add SDDSRVYR
            df_oral_temp.rename(columns={ohapos_col: "OHAPOS"}, inplace=True)
            oral_dfs.append(df_oral_temp)
            print(f"Loaded {url} with OHAPOS column: {ohapos_col}")
        else:
            print(f"OHAPOS column not found in {url}")
    except Exception as e:
        print(f"Could not load {url}. Error: {e}")

# Combine oral health data
if oral_dfs:
    df_oral = pd.concat(oral_dfs, ignore_index=True)
else:
    df_oral = pd.DataFrame()
    print("No OHAPOS data loaded.")

# -----------------------------
# Step 2: Merge all datasets
# -----------------------------
print("Merging all datasets...")
# Use SEQN for merging. Handle potential duplicate column names from merging (e.g., SDDSRVYR)
df = df_demo.merge(df_bmi, on="SEQN", how="left", suffixes=('_demo', '_bmi')) \
            .merge(df_oral, on="SEQN", how="left") \
            .merge(df_physical, on="SEQN", how="left", suffixes=('_oral', '_physical')) \
            .merge(df_drinking, on="SEQN", how="left", suffixes=('_physical', '_drinking')) \
            .merge(df_thyroid, on="SEQN",how="left", suffixes=('_drinking', '_thyroid'))

# Drop redundant SDDSRVYR columns after merging, keeping one or none if all are NaN
# Find all columns starting with 'SDDSRVYR'
sddsrvyr_cols = [col for col in df.columns if col.startswith('SDDSRVYR')]
if sddsrvyr_cols:
    # Keep the first one and drop the rest
    main_sddsrvyr_col = sddsrvyr_cols[0]
    cols_to_drop = [col for col in sddsrvyr_cols if col != main_sddsrvyr_col]
    df.drop(columns=cols_to_drop, inplace=True, errors='ignore')
    df.rename(columns={main_sddsrvyr_col: 'SDDSRVYR'}, inplace=True) # Rename the kept column


# -----------------------------
# Step 3: Classify Variables
# -----------------------------
print("Classifying variables...")

# 1. Classify Free T4 levels (LBXT4F)
def classify_t4f(val):
    if pd.isna(val):
        return np.nan
    elif val < 0.4:
        return "Low T4F (Hypothyroid)"
    elif val <= 4.0:
        return "Normal T4F"
    else:
        return "High T4F (Hyperthyroid)"

df['T4F_Status'] = df['LBXT4F'].apply(classify_t4f)

# 2. Classify Free T3 levels (LBXT3F)
def classify_t3f(val):
    if pd.isna(val):
        return np.nan
    elif val < 2.0:
        return "Low Free T3"
    elif val <= 4.4:
        return "Normal Free T3"
    else:
        return "High Free T3"

df['Free_T3_Status'] = df['LBXT3F'].apply(classify_t3f)


# 3. Classify Sedentary Activity (PAD680)
def sedentary_behavior_pad680(val):
    if pd.isna(val) or val in [7777, 9999]:
        return np.nan
    elif val < 180:
        return "Very active"
    elif 180 <= val < 420:
        return "Active"
    elif val >= 420:
        return "Sedentary"
    return np.nan # Should not be reached with valid input

df['Activity_Level'] = df['PAD680'].apply(sedentary_behavior_pad680)


# 4. Classify BMI (BMXBMI)
def classify_bmi(bmi):
    if pd.isna(bmi):
        return np.nan
    elif bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal weight'
    elif 25 <= bmi < 30:
        return 'Overweight'
    elif 30 <= bmi < 35:
        return 'Obese I'
    elif 35 <= bmi < 40:
        return 'Obese II'
    else:
        return 'Obese III'

df['BMI_Category'] = df['BMXBMI'].apply(classify_bmi)

# 5. Classify Age (RIDAGEYR)
def classify_age(age):
    if pd.isna(age):
        return np.nan
    elif age < 13:
        return 'Child'
    elif 13 <= age < 20:
        return 'Teen'
    elif 20 <= age < 65:
        return 'Adult'
    else:
        return 'Senior'

df['AgeGroup'] = df['RIDAGEYR'].apply(classify_age)

# 6. Classify Drinking Habits (ALQ120Q, ALQ151)
def drinking_category(row):
    freq = row["ALQ120Q"]
    binge = row["ALQ151"]

    # Handle invalid/missing data
    if pd.isna(freq) or freq in [77, 99] or pd.isna(binge) or binge in [7, 9]:
        return np.nan

    # Never drank in the past year
    if freq == 7:
        return "Non-drinker"

    # Binge drinker: drank + had 4/5+ drinks/day at least once
    if binge == 1:
        return "Binge drinker"

    # Frequency-based categories
    if freq in [1, 2]:
        return "Heavy drinker"
    elif freq in [3, 4]:
        return "Moderate drinker"
    elif freq in [5, 6]:
        return "Light drinker"

    return np.nan

df["Drinking habits"] = df.apply(drinking_category, axis=1)

# 7. Map Gender (RIAGENDR)
gender_map = {1.0: 'Male', 2.0: 'Female'}
df['RIAGENDR'] = df['RIAGENDR'].map(gender_map)
# Missing values will become NaN by default using map

# 8. Classify Oral Health Status (OHAPOS)
def classify_oral_health(value):
    if pd.isna(value) or value in [7, 9]: # Assuming 7, 9 are also invalid/unknown codes
        return np.nan # Returning NaN for unknown/missing based on previous steps' pattern
    elif value == 1: # Example: Assuming 1 is 'Good' based on typical coding
        return 'Good oral health'
    elif value == 2: # Example: Assuming 2 is 'Poor'
         return 'Poor oral health'
    else:
        # Handle any other values if necessary based on data dictionary
        return np.nan # Treat other codes as unknown/missing

df['Oral_Health_Status'] = df['OHAPOS'].apply(classify_oral_health)

# 9. Map Survey Years (SDDSRVYR)
cycle_map = {5.0: '2007-2008', 6.0: '2009-2010', 7.0: '2011-2012'}
df['SDDSRVYR'] = df['SDDSRVYR'].map(cycle_map)
# Missing values will become NaN by default using map

# 10. Classify TSH levels
def classify_tsh(val):
    if pd.isna(val):
        return np.nan
    elif val < 0.4:
        return "Low TSH (Hypothyroid)"
    elif val <= 4.0:
        return "Normal TSH"
    else: # val > 4.0
        return "High TSH (Hyperthyroid)"

df['TSH_Status'] = df['LBXTSH1'].apply(classify_tsh)


# -----------------------------
# Step 6: Perform Chi-Squared Tests
# -----------------------------
print("\n--- Performing Chi-Squared Tests for Independence ---")

# Define lists for the categorical predictor and outcome variables (using the classified columns)
categorical_predictors = ['Oral_Health_Status', 'Drinking habits', 'RIAGENDR', 'AgeGroup', 'Activity_Level', 'BMI_Category']
# Corrected categorical outcomes based on classifications
categorical_outcomes = ['T4F_Status', 'TSH_Status', 'Free_T3_Status']


# Iterate through each combination of predictor and outcome variables
for predictor in categorical_predictors:
    # Check if the predictor column exists in the DataFrame
    if predictor not in df.columns:
        print(f"\nWarning: Predictor variable '{predictor}' not found in the DataFrame. Skipping tests for this predictor.")
        continue

    for outcome in categorical_outcomes:
        # Check if the outcome column exists in the DataFrame
        if outcome not in df.columns:
             print(f"\nWarning: Outcome variable '{outcome}' not found in the DataFrame. Skipping test for this outcome.")
             continue

        print(f"\nTesting independence between {predictor} and {outcome}:")

        # Create a contingency table, dropping rows/columns with missing values
        # Use dropna=True to exclude NaN values from the counts
        contingency_table = pd.crosstab(df[predictor], df[outcome], dropna=True)

        # Handle cases where the contingency table is empty after dropping NaNs
        if contingency_table.empty or contingency_table.sum().sum() == 0:
            print("Contingency table is empty or contains no data after dropping NaNs. Skipping test.")
            continue

        # Perform the Chi-Squared test
        try:
            chi2, p, dof, expected = chi2_contingency(contingency_table)

            # Print the results
            print(f"Chi-Squared Statistic: {chi2:.4f}")
            print(f"P-value: {p:.4f}")
            print(f"Degrees of Freedom: {dof}")
            # print("Expected Frequencies:\n", expected) # Optional: print expected frequencies
        except ValueError as e:
             print(f"Could not perform Chi-Squared test for {predictor} and {outcome}. Error: {e}")
        except Exception as e:
             print(f"An unexpected error occurred during Chi-Squared test for {predictor} and {outcome}: {e}")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Assuming the following dataframes are already loaded from previous steps:
# df_physical, df_demo, df_drinking, df_oral, df_thyroid

# -----------------------------
# Step 1: Classify Variables (using the already loaded dataframes)
# -----------------------------
print("Classifying variables from loaded dataframes...")

# 1. Classify Free T3 levels (LBXT3F) from df_thyroid
def classify_t3f(val):
    if pd.isna(val):
        return np.nan
    elif val < 2.0:
        return "Low Free T3"
    elif val <= 4.4:
        return "Normal Free T3"
    else:
        return "High Free T3"

df_thyroid['Free_T3_Status'] = df_thyroid['LBXT3F'].apply(classify_t3f)

# 2. Classify Sedentary Activity (PAD680) from df_physical
def sedentary_behavior_pad680(val):
    if pd.isna(val) or val in [7777, 9999]:
        return np.nan
    elif val < 180:
        return "Very active"
    elif 180 <= val < 420:
        return "Active"
    elif val >= 420:
        return "Sedentary"
    return np.nan # Should not be reached with valid input

df_physical['Activity_Level'] = df_physical['PAD680'].apply(sedentary_behavior_pad680)

# 3. Classify Age (RIDAGEYR) from df_demo
def classify_age(age):
    if pd.isna(age):
        return np.nan
    elif age < 13:
        return 'Child'
    elif 13 <= age < 20:
        return 'Teen'
    elif 20 <= age < 65:
        return 'Adult'
    else:
        return 'Senior'

df_demo['AgeGroup'] = df_demo['RIDAGEYR'].apply(classify_age)

# 4. Classify Drinking Habits (ALQ120Q, ALQ151) from df_drinking
def drinking_category(row):
    freq = row["ALQ120Q"]
    binge = row["ALQ151"]

    # Handle invalid/missing data
    if pd.isna(freq) or freq in [77, 99] or pd.isna(binge) or binge in [7, 9]:
        return np.nan

    # Never drank in the past year
    if freq == 7:
        return "Non-drinker"

    # Binge drinker: drank + had 4/5+ drinks/day at least once
    if binge == 1:
        return "Binge drinker"

    # Frequency-based categories
    if freq in [1, 2]:
        return "Heavy drinker"
    elif freq in [3, 4]:
        return "Moderate drinker"
    elif freq in [5, 6]:
        return "Light drinker"

    return np.nan

df_drinking["Drinking habits"] = df_drinking.apply(drinking_category, axis=1)

# 5. Map Gender (RIAGENDR) from df_demo
gender_map = {1.0: 'Male', 2.0: 'Female'}
df_demo['RIAGENDR'] = df_demo['RIAGENDR'].map(gender_map)
# Missing values will become NaN by default using map

# 6. Classify Oral Health Status (OHAPOS) from df_oral
def classify_oral_health(value):
    if pd.isna(value) or value in [7, 9]: # Assuming 7, 9 are also invalid/unknown codes
        return np.nan # Returning NaN for unknown/missing based on previous steps' pattern
    elif value == 1: # Example: Assuming 1 is 'Good' based on typical coding
        return 'Good oral health'
    elif value == 2: # Example: Assuming 2 is 'Poor'
         return 'Poor oral health'
    else:
        # Handle any other values if necessary based on data dictionary
        return np.nan # Treat other codes as unknown/missing

df_oral['Oral_Health_Status'] = df_oral['OHAPOS'].apply(classify_oral_health)

# 7. Classify TSH levels (LBXTSH1) from df_thyroid
def classify_tsh(val):
    if pd.isna(val):
        return np.nan
    elif val < 0.4:
        return "Low TSH (Hypothyroid)"
    elif val <= 4.0:
        return "Normal TSH"
    else: # val > 4.0
        return "High TSH (Hypothyroid)"

df_thyroid['TSH_Status'] = df_thyroid['LBXTSH1'].apply(classify_tsh)

# -----------------------------
# Step 2: Merge classified dataframes for analysis
# -----------------------------
print("\nMerging classified dataframes for analysis...")

# Select only necessary columns from classified dataframes for merging
df_demo_classified = df_demo[['SEQN', 'RIAGENDR', 'AgeGroup']].copy()
df_physical_classified = df_physical[['SEQN', 'Activity_Level']].copy()
df_drinking_classified = df_drinking[['SEQN', 'Drinking habits']].copy()
df_oral_classified = df_oral[['SEQN', 'Oral_Health_Status']].copy()
df_thyroid_classified = df_thyroid[['SEQN', 'TSH_Status', 'Free_T3_Status']].copy()


# Merge the classified dataframes on SEQN
df_classified = df_demo_classified.merge(df_physical_classified, on='SEQN', how='left') \
                                  .merge(df_drinking_classified, on='SEQN', how='left') \
                                  .merge(df_oral_classified, on='SEQN', how='left') \
                                  .merge(df_thyroid_classified, on='SEQN', how='left')


print(f"Shape of merged classified DataFrame: {df_classified.shape}")
print("\nColumns in merged classified DataFrame:")
print(df_classified.columns.tolist())


# -----------------------------
# Step 3: Perform Chi-Squared Tests
# -----------------------------
print("\n--- Performing Chi-Squared Tests for Independence ---")

# Define lists for the categorical predictor and outcome variables (using the classified columns)
categorical_predictors = ['Oral_Health_Status', 'Drinking habits', 'RIAGENDR', 'AgeGroup', 'Activity_Level']
categorical_outcomes = ['TSH_Status', 'Free_T3_Status']


# Iterate through each combination of predictor and outcome variables
for predictor in categorical_predictors:
    # Check if the predictor column exists in the DataFrame
    if predictor not in df_classified.columns:
        print(f"\nWarning: Predictor variable '{predictor}' not found in the classified DataFrame. Skipping tests for this predictor.")
        continue

    for outcome in categorical_outcomes:
        # Check if the outcome column exists in the DataFrame
        if outcome not in df_classified.columns:
             print(f"\nWarning: Outcome variable '{outcome}' not found in the classified DataFrame. Skipping test for this outcome.")
             continue

        print(f"\nTesting independence between {predictor} and {outcome}:")

        # Create a contingency table, dropping rows/columns with missing values
        # Use dropna=True to exclude NaN values from the counts
        contingency_table = pd.crosstab(df_classified[predictor], df_classified[outcome], dropna=True)

        # Handle cases where the contingency table is empty after dropping NaNs
        if contingency_table.empty or contingency_table.sum().sum() == 0:
            print("Contingency table is empty or contains no data after dropping NaNs. Skipping test.")
            continue

        # Perform the Chi-Squared test
        try:
            chi2, p, dof, expected = chi2_contingency(contingency_table)

            # Print the results
            print(f"Chi-Squared Statistic: {chi2:.4f}")
            print(f"P-value: {p:.4f}")
            print(f"Degrees of Freedom: {dof}")
            # print("Expected Frequencies:\n", expected) # Optional: print expected frequencies
        except ValueError as e:
             print(f"Could not perform Chi-Squared test for {predictor} and {outcome}. Error: {e}")
        except Exception as e:
             print(f"An unexpected error occurred during Chi-Squared test for {predictor} and {outcome}: {e}")
```

```python
import pandas as pd
import numpy as np
import zipfile
import os
from scipy.stats import chi2_contingency

# -----------------------------
# Step 0: Define zip file path and extraction directory
# -----------------------------
# Corrected zip file path based on user input
zip_file_path = "/content/Biopal proj zip.zip"
extract_dir = "/content/extracted_nhanes_data"

# Create the extraction directory if it doesn't exist
os.makedirs(extract_dir, exist_ok=True)

# Unzip the file
# Add error handling for the zip file
try:
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extract_dir)
    print(f"Successfully extracted data to {extract_dir}")
except FileNotFoundError:
    print(f"Error: The zip file '{zip_file_path}' was not found.")
    # Exit or handle the error as appropriate for your script
    exit()
except zipfile.BadZipFile:
    print(f"Error: The file '{zip_file_path}' is not a valid zip file.")
    # Exit or handle the error
    exit()
except Exception as e:
    print(f"An unexpected error occurred during zip extraction: {e}")
    # Exit or handle the error
    exit()


# -----------------------------
# Step 1: Define file prefixes and columns and Create load_data function
# -----------------------------
print("Defining file prefixes and columns and creating load_data function...")

# Dictionary mapping file prefixes to columns to load
file_column_map = {
    "DEMO": ["SEQN", "SDDSRVYR", "RIAGENDR", "RIDAGEYR"],
    "BMX": ["SEQN", "SDDSRVYR", "BMXBMI"],
    "PAQ": ["SEQN", "SDDSRVYR", "PAD680"],
    "ALQ": ["SEQN", "SDDSRVYR", "ALQ120Q", "ALQ151"],
    "THYROD": ["SEQN", "SDDSRVYR", "LBXTSH1", "LBXT4F", "LBXT3F"],
}

def load_data(file_prefix, columns):
    df_list = []
    # List all files in the extracted directory that match the file_prefix
    # Handle potential duplicated files with numbers in parentheses
    extracted_files = [os.path.join(extract_dir, f) for f in os.listdir(extract_dir)
                       if f.startswith(file_prefix) and f.endswith('.xpt')]

    if not extracted_files:
        print(f"No files found for prefix {file_prefix} in {extract_dir}")
        return pd.DataFrame()

    for file_path in extracted_files:
        try:
            df_temp = pd.read_sas(file_path)
            existing_cols = [col for col in columns if col in df_temp.columns]
            if existing_cols:
                df_list.append(df_temp[existing_cols])
                print(f"Loaded {file_path} with columns: {existing_cols}")
            else:
                print(f"No matching columns found in {file_path}")
        except Exception as e:
            print(f"Could not load or process {file_path}. Error: {e}")
    if df_list:
        return pd.concat(df_list, ignore_index=True)
    else:
        print(f"No data loaded for {file_prefix}. Returning empty DataFrame.")
        return pd.DataFrame()

# -----------------------------
# Step 2: Load and process Oral Health (OHX/OHXREF) data
# -----------------------------
print("Loading and processing OHAPOS data from OHX and OHXREF...")

# Each tuple: (year, cycle, file_prefix, SDDSRVYR value)
oral_files = [
    ("2007", "E", "OHX", 5),
    ("2009", "F", "OHXREF", 6),
    ("2011", "G", "OHXREF", 7)
]

oral_dfs = []
for year, cycle, prefix, survey_cycle in oral_files:
    # Construct the expected filename based on prefix and cycle
    file_name = f"{prefix}_{cycle}.xpt"
    file_path = os.path.join(extract_dir, file_name)

    # Also check for files with numbers in parentheses like "OHX_E (1).xpt"
    potential_files = [f for f in os.listdir(extract_dir) if f.startswith(f"{prefix}_{cycle}")]
    found_file = None
    for pf in potential_files:
        if pf.endswith('.xpt'):
            found_file = os.path.join(extract_dir, pf)
            break

    if not found_file:
        print(f"File not found for {prefix}_{cycle}.xpt in {extract_dir}")
        continue

    try:
        df_oral_temp = pd.read_sas(found_file)
        # Find OHAPOS column case-insensitively
        ohapos_col = next((col for col in df_oral_temp.columns if "OHAPOS" in col.upper()), None)
        if ohapos_col:
            # Select SEQN and the found OHAPOS column
            df_oral_temp = df_oral_temp[["SEQN", ohapos_col]].copy()
            df_oral_temp["SDDSRVYR"] = survey_cycle  # Manually add SDDSRVYR
            df_oral_temp.rename(columns={ohapos_col: "OHAPOS"}, inplace=True)
            oral_dfs.append(df_oral_temp)
            print(f"Loaded {found_file} with OHAPOS column: {ohapos_col}")
        else:
            print(f"OHAPOS column not found in {found_file}")
    except Exception as e:
        print(f"Could not load {found_file}. Error: {e}")

# Combine oral health data
if oral_dfs:
    df_oral = pd.concat(oral_dfs, ignore_index=True)
else:
    df_oral = pd.DataFrame()
    print("No OHAPOS data loaded.")


# -----------------------------
# Step 3: Load other datasets
# -----------------------------
print("Loading other datasets...")
df_demo = load_data("DEMO", file_column_map["DEMO"])
df_bmi = load_data("BMX", file_column_map["BMX"])
df_physical = load_data("PAQ", file_column_map["PAQ"])
df_drinking = load_data("ALQ", file_column_map["ALQ"])
df_thyroid = load_data("THYROD", file_column_map["THYROD"])

# -----------------------------
# Step 4: Merge all datasets
# -----------------------------
print("Merging all datasets...")
# Use SEQN for merging. Handle potential duplicate column names from merging (e.g., SDDSRVYR)
df = df_demo.merge(df_bmi, on="SEQN", how="left", suffixes=('_demo', '_bmi')) \
            .merge(df_oral, on="SEQN", how="left") \
            .merge(df_physical, on="SEQN", how="left", suffixes=('_oral', '_physical')) \
            .merge(df_drinking, on="SEQN", how="left", suffixes=('_physical', '_drinking')) \
            .merge(df_thyroid, on="SEQN",how="left", suffixes=('_drinking', '_thyroid'))

# Drop redundant SDDSRVYR columns after merging, keeping one or none if all are NaN
# Find all columns starting with 'SDDSRVYR'
sddsrvyr_cols = [col for col in df.columns if col.startswith('SDDSRVYR')]
if sddsrvyr_cols:
    # Keep the first one and drop the rest
    main_sddsrvyr_col = sddsrvyr_cols[0]
    cols_to_drop = [col for col in sddsrvyr_cols if col != main_sddsrvyr_col]
    df.drop(columns=cols_to_drop, inplace=True, errors='ignore')
    df.rename(columns={main_sddsrvyr_col: 'SDDSRVYR'}, inplace=True) # Rename the kept column


# -----------------------------
# Step 5: Classify Variables
# -----------------------------
print("Classifying variables...")

# 1. Classify Free T4 levels (LBXT4F)
def classify_t4f(val):
    if pd.isna(val):
        return np.nan
    elif val < 0.4:
        return "Low T4F (Hypothyroid)"
    elif val <= 4.0:
        return "Normal T4F"
    else:
        return "High T4F (Hyperthyroid)"

df['T4F_Status'] = df['LBXT4F'].apply(classify_t4f)

# 2. Classify Free T3 levels (LBXT3F)
def classify_t3f(val):
    if pd.isna(val):
        return np.nan
    elif val < 2.0:
        return "Low Free T3"
    elif val <= 4.4:
        return "Normal Free T3"
    else:
        return "High Free T3"

df['Free_T3_Status'] = df['LBXT3F'].apply(classify_t3f)


# 3. Classify Sedentary Activity (PAD680)
def sedentary_behavior_pad680(val):
    if pd.isna(val) or val in [7777, 9999]:
        return np.nan
    elif val < 180:
        return "Very active"
    elif 180 <= val < 420:
        return "Active"
    elif val >= 420:
        return "Sedentary"
    return np.nan # Should not be reached with valid input

df['Activity_Level'] = df['PAD680'].apply(sedentary_behavior_pad680)


# 4. Classify BMI (BMXBMI)
def classify_bmi(bmi):
    if pd.isna(bmi):
        return np.nan
    elif bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal weight'
    elif 25 <= bmi < 30:
        return 'Overweight'
    elif 30 <= bmi < 35:
        return 'Obese I'
    elif 35 <= bmi < 40:
        return 'Obese II'
    else:
        return 'Obese III'

df['BMI_Category'] = df['BMXBMI'].apply(classify_bmi)

# 5. Classify Age (RIDAGEYR)
def classify_age(age):
    if pd.isna(age):
        return np.nan
    elif age < 13:
        return 'Child'
    elif 13 <= age < 20:
        return 'Teen'
    elif 20 <= age < 65:
        return 'Adult'
    else:
        return 'Senior'

df['AgeGroup'] = df['RIDAGEYR'].apply(classify_age)

# 6. Classify Drinking Habits (ALQ120Q, ALQ151)
def drinking_category(row):
    freq = row["ALQ120Q"]
    binge = row["ALQ151"]

    # Handle invalid/missing data
    if pd.isna(freq) or freq in [77, 99] or pd.isna(binge) or binge in [7, 9]:
        return np.nan

    # Never drank in the past year
    if freq == 7:
        return "Non-drinker"

    # Binge drinker: drank + had 4/5+ drinks/day at least once
    if binge == 1:
        return "Binge drinker"

    # Frequency-based categories
    if freq in [1, 2]:
        return "Heavy drinker"
    elif freq in [3, 4]:
        return "Moderate drinker"
    elif freq in [5, 6]:
        return "Light drinker"

    return np.nan

df["Drinking habits"] = df.apply(drinking_category, axis=1)

# 7. Map Gender (RIAGENDR)
gender_map = {1.0: 'Male', 2.0: 'Female'}
df['RIAGENDR'] = df['RIAGENDR'].map(gender_map)
# Missing values will become NaN by default using map

# 8. Classify Oral Health Status (OHAPOS)
def classify_oral_health(value):
    if pd.isna(value) or value in [7, 9]: # Assuming 7, 9 are also invalid/unknown codes
        return np.nan # Returning NaN for unknown/missing based on previous steps' pattern
    elif value == 1: # Example: Assuming 1 is 'Good' based on typical coding
        return 'Good oral health'
    elif value == 2: # Example: Assuming 2 is 'Poor'
         return 'Poor oral health'
    else:
        # Handle any other values if necessary based on data dictionary
        return np.nan # Treat other codes as unknown/missing

df['Oral_Health_Status'] = df['OHAPOS'].apply(classify_oral_health)

# 9. Map Survey Years (SDDSRVYR)
cycle_map = {5.0: '2007-2008', 6.0: '2009-2010', 7.0: '2011-2012'}
df['SDDSRVYR'] = df['SDDSRVYR'].map(cycle_map)
# Missing values will become NaN by default using map

# 10. Classify TSH levels
def classify_tsh(val):
    if pd.isna(val):
        return np.nan
    elif val < 0.4:
        return "Low TSH (Hypothyroid)"
    elif val <= 4.0:
        return "Normal TSH"
    else: # val > 4.0
        return "High TSH (Hypothyroid)"

df['TSH_Status'] = df['LBXTSH1'].apply(classify_tsh)


# -----------------------------
# Step 6: Perform Chi-Squared Tests
# -----------------------------
print("\n--- Performing Chi-Squared Tests for Independence ---")

# Define lists for the categorical predictor and outcome variables (using the classified columns)
categorical_predictors = ['Oral_Health_Status', 'Drinking habits', 'RIAGENDR', 'AgeGroup', 'Activity_Level', 'BMI_Category']
# Corrected categorical outcomes based on classifications
categorical_outcomes = ['T4F_Status', 'TSH_Status', 'Free_T3_Status']


# Iterate through each combination of predictor and outcome variables
for predictor in categorical_predictors:
    # Check if the predictor column exists in the DataFrame
    if predictor not in df.columns:
        print(f"\nWarning: Predictor variable '{predictor}' not found in the DataFrame. Skipping tests for this predictor.")
        continue

    for outcome in categorical_outcomes:
        # Check if the outcome column exists in the DataFrame
        if outcome not in df.columns:
             print(f"\nWarning: Outcome variable '{outcome}' not found in the DataFrame. Skipping test for this outcome.")
             continue

        print(f"\nTesting independence between {predictor} and {outcome}:")

        # Create a contingency table, dropping rows/columns with missing values
        # Use dropna=True to exclude NaN values from the counts
        contingency_table = pd.crosstab(df[predictor], df[outcome], dropna=True)

        # Handle cases where the contingency table is empty after dropping NaNs
        if contingency_table.empty or contingency_table.sum().sum() == 0:
            print("Contingency table is empty or contains no data after dropping NaNs. Skipping test.")
            continue

        # Perform the Chi-Squared test
        try:
            chi2, p, dof, expected = chi2_contingency(contingency_table)

            # Print the results
            print(f"Chi-Squared Statistic: {chi2:.4f}")
            print(f"P-value: {p:.4f}")
            print(f"Degrees of Freedom: {dof}")
            # print("Expected Frequencies:\n", expected) # Optional: print expected frequencies
        except ValueError as e:
             print(f"Could not perform Chi-Squared test for {predictor} and {outcome}. Error: {e}")
        except Exception as e:
             print(f"An unexpected error occurred during Chi-Squared test for {predictor} and {outcome}: {e}")
```

```python
import pandas as pd
import numpy as np
import zipfile
import os
from scipy.stats import chi2_contingency # Import chi2_contingency as it was used before and might be needed later

# -----------------------------
# Step 0: Define zip file path and extraction directory
# -----------------------------
# Corrected zip file path based on user input
zip_file_path = "/content/Biopal proj zip.zip"
extract_dir = "/content/extracted_nhanes_data"

# Create the extraction directory if it doesn't exist
os.makedirs(extract_dir, exist_ok=True)

# Unzip the file
# Add error handling for the zip file
try:
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extract_dir)
    print(f"Successfully extracted data to {extract_dir}")
except FileNotFoundError:
    print(f"Error: The zip file '{zip_file_path}' was not found.")
    # Exit or handle the error as appropriate for your script
    exit()
except zipfile.BadZipFile:
    print(f"Error: The file '{zip_file_path}' is not a valid zip file.")
    # Exit or handle the error
    exit()
except Exception as e:
    print(f"An unexpected error occurred during zip extraction: {e}")
    # Exit or handle the error
    exit()


# -----------------------------
# Step 1: Define file prefixes and columns and Create load_data function
# -----------------------------
print("Defining file prefixes and columns and creating load_data function...")

# Dictionary mapping file prefixes to columns to load
file_column_map = {
    "DEMO": ["SEQN", "SDDSRVYR", "RIAGENDR", "RIDAGEYR"],
    "BMX": ["SEQN", "SDDSRVYR", "BMXBMI"],
    "PAQ": ["SEQN", "SDDSRVYR", "PAD680"],
    "ALQ": ["SEQN", "SDDSRVYR", "ALQ120Q", "ALQ151"],
    "THYROD": ["SEQN", "SDDSRVYR", "LBXTSH1", "LBXT4F", "LBXT3F"],
}

def load_data(file_prefix, columns):
    df_list = []
    # List all files in the extracted directory that match the file_prefix
    # Handle potential duplicated files with numbers in parentheses
    extracted_files = [os.path.join(extract_dir, f) for f in os.listdir(extract_dir)
                       if f.startswith(file_prefix) and f.endswith('.xpt')]

    if not extracted_files:
        print(f"No files found for prefix {file_prefix} in {extract_dir}")
        return pd.DataFrame()

    for file_path in extracted_files:
        try:
            df_temp = pd.read_sas(file_path)
            existing_cols = [col for col in columns if col in df_temp.columns]
            if existing_cols:
                df_list.append(df_temp[existing_cols])
                print(f"Loaded {file_path} with columns: {existing_cols}")
            else:
                print(f"No matching columns found in {file_path}")
        except Exception as e:
            print(f"Could not load or process {file_path}. Error: {e}")
    if df_list:
        return pd.concat(df_list, ignore_index=True)
    else:
        print(f"No data loaded for {file_prefix}. Returning empty DataFrame.")
        return pd.DataFrame()

# -----------------------------
# Step 2: Load and process Oral Health (OHX/OHXREF) data
# -----------------------------
print("Loading and processing OHAPOS data from OHX and OHXREF...")

# Each tuple: (year, cycle, file_prefix, SDDSRVYR value)
oral_files = [
    ("2007", "E", "OHX", 5),
    ("2009", "F", "OHXREF", 6),
    ("2011", "G", "OHXREF", 7)
]

oral_dfs = []
for year, cycle, prefix, survey_cycle in oral_files:
    # Construct the expected filename based on prefix and cycle
    file_name = f"{prefix}_{cycle}.xpt"
    file_path = os.path.join(extract_dir, file_name)

    # Also check for files with numbers in parentheses like "OHX_E (1).xpt"
    potential_files = [f for f in os.listdir(extract_dir) if f.startswith(f"{prefix}_{cycle}")]
    found_file = None
    for pf in potential_files:
        if pf.endswith('.xpt'):
            found_file = os.path.join(extract_dir, pf)
            break

    if not found_file:
        print(f"File not found for {prefix}_{cycle}.xpt in {extract_dir}")
        continue

    try:
        df_oral_temp = pd.read_sas(found_file)
        # Find OHAPOS column case-insensitively
        ohapos_col = next((col for col in df_oral_temp.columns if "OHAPOS" in col.upper()), None)
        if ohapos_col:
            # Select SEQN and the found OHAPOS column
            df_oral_temp = df_oral_temp[["SEQN", ohapos_col]].copy()
            df_oral_temp["SDDSRVYR"] = survey_cycle  # Manually add SDDSRVYR
            df_oral_temp.rename(columns={ohapos_col: "OHAPOS"}, inplace=True)
            oral_dfs.append(df_oral_temp)
            print(f"Loaded {found_file} with OHAPOS column: {ohapos_col}")
        else:
            print(f"OHAPOS column not found in {found_file}")
    except Exception as e:
        print(f"Could not load {found_file}. Error: {e}")

# Combine oral health data
if oral_dfs:
    df_oral = pd.concat(oral_dfs, ignore_index=True)
else:
    df_oral = pd.DataFrame()
    print("No OHAPOS data loaded.")


# -----------------------------
# Step 3: Load other datasets
# -----------------------------
print("Loading other datasets...")
df_demo = load_data("DEMO", file_column_map["DEMO"])
df_bmi = load_data("BMX", file_column_map["BMX"])
df_physical = load_data("PAQ", file_column_map["PAQ"])
df_drinking = load_data("ALQ", file_column_map["ALQ"])
df_thyroid = load_data("THYROD", file_column_map["THYROD"])

# -----------------------------
# Step 4: Merge all datasets
# -----------------------------
print("Merging all datasets...")
# Use SEQN for merging. Handle potential duplicate column names from merging (e.g., SDDSRVYR)
df = df_demo.merge(df_bmi, on="SEQN", how="left", suffixes=('_demo', '_bmi')) \
            .merge(df_oral, on="SEQN", how="left") \
            .merge(df_physical, on="SEQN", how="left", suffixes=('_oral', '_physical')) \
            .merge(df_drinking, on="SEQN", how="left", suffixes=('_physical', '_drinking')) \
            .merge(df_thyroid, on="SEQN",how="left", suffixes=('_drinking', '_thyroid'))

# Drop redundant SDDSRVYR columns after merging, keeping one or none if all are NaN
# Find all columns starting with 'SDDSRVYR'
sddsrvyr_cols = [col for col in df.columns if col.startswith('SDDSRVYR')]
if sddsrvyr_cols:
    # Keep the first one and drop the rest
    main_sddsrvyr_col = sddsrvyr_cols[0]
    cols_to_drop = [col for col in sddsrvyr_cols if col != main_sddsrvyr_col]
    df.drop(columns=cols_to_drop, inplace=True, errors='ignore')
    df.rename(columns={main_sddsrvyr_col: 'SDDSRVYR'}, inplace=True) # Rename the kept column


# -----------------------------
# Step 5: Classify Variables (Re-classify after reloading data)
# -----------------------------
print("Classifying variables after merging...")

# 1. Classify Free T4 levels (LBXT4F)
def classify_t4f(val):
    if pd.isna(val):
        return np.nan
    elif val < 0.4:
        return "Low T4F (Hypothyroid)"
    elif val <= 4.0:
        return "Normal T4F"
    else:
        return "High T4F (Hyperthyroid)"

df['T4F_Status'] = df['LBXT4F'].apply(classify_t4f)

# 2. Classify Free T3 levels (LBXT3F)
def classify_t3f(val):
    if pd.isna(val):
        return np.nan
    elif val < 2.0:
        return "Low Free T3"
    elif val <= 4.4:
        return "Normal Free T3"
    else:
        return "High Free T3"

df['Free_T3_Status'] = df['LBXT3F'].apply(classify_t3f)


# 3. Classify Sedentary Activity (PAD680)
def sedentary_behavior_pad680(val):
    if pd.isna(val) or val in [7777, 9999]:
        return np.nan
    elif val < 180:
        return "Very active"
    elif 180 <= val < 420:
        return "Active"
    elif val >= 420:
        return "Sedentary"
    return np.nan # Should not be reached with valid input

df['Activity_Level'] = df['PAD680'].apply(sedentary_behavior_pad680)


# 4. Classify BMI (BMXBMI)
def classify_bmi(bmi):
    if pd.isna(bmi):
        return np.nan
    elif bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal weight'
    elif 25 <= bmi < 30:
        return 'Overweight'
    elif 30 <= bmi < 35:
        return 'Obese I'
    elif 35 <= bmi < 40:
        return 'Obese II'
    else:
        return 'Obese III'

df['BMI_Category'] = df['BMXBMI'].apply(classify_bmi)

# 5. Classify Age (RIDAGEYR)
def classify_age(age):
    if pd.isna(age):
        return np.nan
    elif age < 13:
        return 'Child'
    elif 13 <= age < 20:
        return 'Teen'
    elif 20 <= age < 65:
        return 'Adult'
    else:
        return 'Senior'

df['AgeGroup'] = df['RIDAGEYR'].apply(classify_age)

# 6. Classify Drinking Habits (ALQ120Q, ALQ151)
def drinking_category(row):
    freq = row["ALQ120Q"]
    binge = row["ALQ151"]

    # Handle invalid/missing data
    if pd.isna(freq) or freq in [77, 99] or pd.isna(binge) or binge in [7, 9]:
        return np.nan

    # Never drank in the past year
    if freq == 7:
        return "Non-drinker"

    # Binge drinker: drank + had 4/5+ drinks/day at least once
    if binge == 1:
        return "Binge drinker"

    # Frequency-based categories
    if freq in [1, 2]:
        return "Heavy drinker"
    elif freq in [3, 4]:
        return "Moderate drinker"
    elif freq in [5, 6]:
        return "Light drinker"

    return np.nan

df["Drinking habits"] = df.apply(drinking_category, axis=1)

# 7. Map Gender (RIAGENDR)
gender_map = {1.0: 'Male', 2.0: 'Female'}
df['RIAGENDR'] = df['RIAGENDR'].map(gender_map)
# Missing values will become NaN by default using map

# 8. Classify Oral Health Status (OHAPOS)
def classify_oral_health(value):
    if pd.isna(value) or value in [7, 9]: # Assuming 7, 9 are also invalid/unknown codes
        return np.nan # Returning NaN for unknown/missing based on previous steps' pattern
    elif value == 1: # Example: Assuming 1 is 'Good' based on typical coding
        return 'Good oral health'
    elif value == 2: # Example: Assuming 2 is 'Poor'
         return 'Poor oral health'
    else:
        # Handle any other values if necessary based on data dictionary
        return np.nan # Treat other codes as unknown/missing

df['Oral_Health_Status'] = df['OHAPOS'].apply(classify_oral_health)

# 9. Map Survey Years (SDDSRVYR)
cycle_map = {5.0: '2007-2008', 6.0: '2009-2010', 7.0: '2011-2012'}
df['SDDSRVYR'] = df['SDDSRVYR'].map(cycle_map)
# Missing values will become NaN by default using map

# 10. Classify TSH levels
def classify_tsh(val):
    if pd.isna(val):
        return np.nan
    elif val < 0.4:
        return "Low TSH (Hypothyroid)"
    elif val <= 4.0:
        return "Normal TSH"
    else: # val > 4.0
        return "High TSH (Hypothyroid)"

df['TSH_Status'] = df['LBXTSH1'].apply(classify_tsh)


# -----------------------------
# Step 6: Perform Multiple Linear Regression
# -----------------------------
import statsmodels.api as sm
from statsmodels.formula.api import ols


# Define the continuous and categorical predictor variables for regression
continuous_predictors_reg = ['BMXBMI', 'PAD680', 'RIDAGEYR'] # BMI, Physical Activity (Sedentary Mins), Age
categorical_predictors_reg = ['RIAGENDR', 'Drinking habits', 'Oral_Health_Status', 'Activity_Level', 'BMI_Category', 'AgeGroup']

# Define the continuous thyroid outcome variables for regression
continuous_outcomes_reg = ['LBXT4F', 'LBXT3F', 'LBXTSH1'] # Free T4, Free T3, TSH


# Filter the DataFrame to include only relevant columns and drop rows with any missing values
# Dropping NaNs is necessary for linear regression.
# We need to include all predictors and all outcomes for this analysis.
analysis_cols_regression = continuous_predictors_reg + categorical_predictors_reg + continuous_outcomes_reg
df_analysis_regression = df[analysis_cols_regression].dropna().copy()


print(f"\nShape of DataFrame for Multiple Linear Regression analysis after dropping NaNs: {df_analysis_regression.shape}")
print("\nColumns used in Multiple Linear Regression analysis:")
print(df_analysis_regression.columns.tolist())


# Prepare predictor variables (X) for Multiple Linear Regression
# We need to handle categorical variables by creating dummy variables
X_categorical_regression = df_analysis_regression[categorical_predictors_reg]
X_categorical_dummies_regression = pd.get_dummies(X_categorical_regression, columns=categorical_predictors_reg, drop_first=True) # Drop_first=True to avoid multicollinearity

# Combine continuous and dummy variables to form the predictor matrix X
X_regression = pd.concat([df_analysis_regression[continuous_predictors_reg], X_categorical_dummies_regression], axis=1)

# Add a constant term to the predictor matrix X
X_regression = sm.add_constant(X_regression)

print("\nShape of X_regression (predictor matrix):", X_regression.shape)
print("\nFirst 5 rows of X_regression:")
display(X_regression.head())


# Perform Multiple Linear Regression for each continuous thyroid outcome
print("\n--- Performing Multiple Linear Regression Analysis ---")

for outcome in continuous_outcomes_reg:
     if outcome in df_analysis_regression.columns:
        y_regression = df_analysis_regression[outcome]
        print(f"\nMultiple Linear Regression for {outcome} as outcome:")
        try:
            # Fit the OLS model
            model = sm.OLS(y_regression, X_regression).fit()
            # Print the summary
            print(model.summary())
        except Exception as e:
            print(f"Could not perform regression for {outcome}. Error: {e}")
     else:
        print(f"Outcome '{outcome}' not found in analysis DataFrame. Skipping regression for this outcome.")
```
